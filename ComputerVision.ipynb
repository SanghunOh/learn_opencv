```
import pandas as pd
df_one = pd.read_excel('./handMNIST2828_one_predict.xlsx', header=None)
import matplotlib.pyplot as plt
plt.imshow(df_one.values, cmap='gray')
```

```
df_two = pd.read_excel('./handMNIST2828_two_predict.xlsx', header=None)
plt.imshow(df_two.values, cmap='gray')
```

```
df_matrix = pd.read_excel('./handMNIST2828_flipMatrix.xlsx', header=None)
plt.imshow(df_matrix.values, cmap='gray')
```

```
import numpy as np
```

```
np_two = df_two.to_numpy()
np_matrix = df_matrix.to_numpy()
type(np_two), np_two.shape, type(np_matrix), np_matrix.shape
```

```
np_hflip = np.matmul(np_two, np_matrix)
type(np_hflip), np_hflip.shape
```

```
df_hflip = pd.DataFrame(data=np_hflip)
plt.imshow(df_hflip.values, cmap='gray')
```

# Numpy with Grayscale

```
import numpy as np
from PIL import Image

# gradient between 0 and 1 for 256*256
array = np.linspace(0,1,256*256)

# reshape to 2d
mat = np.reshape(array,(256,256))

# Creates PIL image
img = Image.fromarray(np.uint8(mat * 255) , 'L')
img.show()
```
